# Combining and Merging Datasets

```python
fileData=pd.read_csv('iskandarAnis.csv')
type(fileData)
```

```python
Data=pd.DataFrame(pd.read_csv('iskandarAnis.csv'))
type(Data)
```

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'], 'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],'data2': range(3)})
df1 
```

```python
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='key')
```

# OUTER JOIN

```python
mergedOuter=pd.merge(df1,df2,how='outer',on='key')
mergedOuter
```

```python
mergedOuter=pd.merge(df1,df2,how='outer',on='key')
mergedOuter
```

# LEFT JOIN

```python
mergedLeft=pd.merge(df1,df2,how='left',on='key')
mergedLeft
```

```python
mergedLeft_=pd.merge(df2,df1,how='left', on='key')
mergedLeft_
```

# RIGHT JOIN

```python
mergedRight=pd.merge(df1,df2,how='right',on='key')
mergedRight
```

# MERGING ON INDEX

```python
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'], 'value': range(6)})
right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
left1 
```

```python
right1
```

```python
pd.merge(left1,right1,left_on='key',right_index=True)
```

# CONCATENATING ALONG AN AXIS

```python
studentFrame=pd.DataFrame({'Name':['iskandar','Humaira','Akmal','Hilman'],
                                                'Grade':['A+','A+','A-','B+']})
studentFrame
```

```python
studentFrame_=pd.DataFrame({"Name":['Afiq','Syahmi'],'Grade':['B','C-']})
studentFrame_
```

```python
pd.concat([studentFrame,studentFrame_])
```
