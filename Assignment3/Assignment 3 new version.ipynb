```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('Diabetes Dataset.csv')
data
```

```python
data.shape
```

```python
columnMissingDF=[]
for column in data:
    x=data[column].isnull().sum()
    if(x>0):
        columnMissingDF.append({'column':column,'missingCount':x})
        
missing=pd.DataFrame(columnMissingDF)
missing
```

```python
data=data.fillna({'No_Times_Pregnant':1})
data
```

```python
data=data.dropna(thresh=4)
data
```

```python
data=data.fillna({'Plasma_Glucose':data['Plasma_Glucose'].mean(),
                          'Diastolic':data['Diastolic'].fillna(method='ffill'),
                           'Triceps':data['Triceps'].fillna(method='bfill'),
                           'Insulin':data['Insulin'].mean(),
                           'BMI':data['BMI'].median()})
data
```

```python
columnMissingDF=[]
for column in data:
    x=data[column].isnull().sum()
    if(x>0):
        columnMissingDF.append({'column':column,'missingCount':x})
        
missing=pd.DataFrame(columnMissingDF)
missing
```

```python
ages=data['Age']
ages
```

```python
bins=[x for x in range(20,100,10)]
bins
```

```python
grouped=pd.cut(ages,bins,right=False)
pd.value_counts(grouped)
```
