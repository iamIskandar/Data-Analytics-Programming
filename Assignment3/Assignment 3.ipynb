# Task 1

```python
#TASK 1
# 1. Read the "Diabetes Dataset.csv" and save the data into a DataFrame
diabetes_frame=pd.read_csv('Diabetes Dataset.csv')
diabetes_frame
```

```python
# 2. What is the shape of the DataFrame
print('The shape of the DataFrame is',diabetes_frame.shape)
```

```python
#3. Get the number of missing values for each column
print("the number of missing values for each column are as follows")
diabetes_frame.isna().sum()
```

```python
#4. Fill in the missing values in column No_Times_Pregnant with value 1.
diabetes_frame['No_Times_Pregnant']=diabetes_frame['No_Times_Pregnant'].fillna(1)
diabetes_frame
```

```python
#5. Drop all rows which contains less than 4 observation values
diabetes_frame=diabetes_frame.dropna(thresh=4)
diabetes_frame
```

```python
#6 Fill in all the missing values in below columns with the given values or methods
copyFrame=diabetes_frame.copy()
copyFrame.fillna({'Plasma_Glucose':diabetes_frame['Plasma_Glucose'].mean(),
                       'Diastolic':diabetes_frame['Diastolic'].fillna(method='ffill'),
                       'Triceps':diabetes_frame['Triceps'].fillna(method='bfill'),
                       'Insulin':diabetes_frame['Insulin'].mean(),
                       'BMI':diabetes_frame['BMI'].median()},inplace=True)
#copyFrame.head(20)
diabetes_frame=copyFrame
diabetes_frame.head()
```

```python
#7. Check back the number fo missing values for each column. It should now be 0 for all the columns
diabetes_frame.isna().sum()
```

```python
#8. Bin the age into below categories. How many patients with diabetes for each category
ages=diabetes_frame['Age']
bins=[20,30,40,50,60,70,80,90]
group=pd.cut(ages,bins,right=False)
pd.value_counts(group)
```

# Task 2

```python
#1. Read the "Live births by state and sex.xlsx" and save the data into a DataFrame
birthInfo=pd.read_excel('Live births by state and sex.xlsx')
birthInfo
```

```python
#2. Read another file named 'Death by state and sex.xlsx' and save it into another DataFrame
deathInfo=pd.read_excel('Death by state and sex.xlsx')
deathInfo
```

```python
#3. Merge Both files.
fullInfo=pd.merge(birthInfo,deathInfo, on=['Year','State','Sex'])
fullInfo
```

```python
#4. Set the index as figure below
fullInfo=fullInfo.set_index(['Year','State'])
fullInfo
```

```python
#5. Display the number of live births and number of deaths for all the years (2009-2018)
groupedInfo=fullInfo.sum(level='Year').sort_index(ascending=False)
groupedInfo
```

```python
#6 .Swap the level as figure below
fullInfo.swaplevel('Year','State').sort_index(level=0)
```

```python
#7. Display the number of live births and number of deaths for all states as below
fullInfo.sum(level='State')
```
