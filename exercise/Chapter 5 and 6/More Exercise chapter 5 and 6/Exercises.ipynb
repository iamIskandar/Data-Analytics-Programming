### Exercise 2
Given two sets of data about restaurant visit in Japan:<br>
<b>air_store_info.csv</b> - store's information from AirREGI / Restaurant Board similar to Square.<br> 
<b>air_visit_data.csv</b> - the visit data information to each of the store (refer store ID), date and amount of visitors.<br>
<b>date_info.csv</b>      - the information of the date and holiday flag (if the day is holiday, flag = 1)<br>

The dataset is part of the datasets shared in: https://www.kaggle.com/c/recruit-restaurant-visitor-forecasting/data

1. Read all the datasets separately using the `read_csv` function in pandas

```python
store_info = pd.read_csv('air_store_info.csv')
visit_data = pd.read_csv('air_visit_data.csv')
date_info = pd.read_csv('date_info.csv')
```

2. Create a new data frame by merging the data from `air_visit_data.csv` to `air_store_info.csv` to get the corresponding information regarding the store.

```python
merged=pd.merge(store_info,visit_data,on='air_store_id',how='left')
merged
```

3. Using the resulting data frame from (2), merge with `date_info.csv` based on the key `date` into a new data frame that consists of the holiday flag information. 

```python
merged_=pd.merge(merged,date_info,left_on='visit_date',right_on='calendar_date',how='left')
merged_
```

4. Check the new data frame from (3) if there is any missing values. If yes, remove the data entries (rows) with missing values

```python
columnMissingDF=[]
for column in merged_:
    x=merged_[column].isnull().sum()
    columnMissingDF.append({'column':column,'missingCount':x})

missing=pd.DataFrame(columnMissingDF)
missing
```
