# Data Aggregration and group operations

```python
#displaying the total number of customers registered in this dataset
print('Total number of customers from both hotels categories =>',data['customers'].count(),'people')
```

```python
#Determine the number of customers from cityy and resort hotel using groupby() method
print('Number of customers from City and Resort hotel')
data[['customers']].groupby(data['hotel']).count()
```

```python
#Determne the number of customers from city and resort hotel based on year using groupby() meethpd
print('Number of customers from City and Resort hotel based on year')
grouped=data[['customers']].groupby([data['hotel'],data['year']]).count()
grouped
```

```python
#Performing aggregation operation to determine the analytics for week night stays in and weekend night stays in usinng groupby() method and 
#aggregate functions 
aggMethods=['sum','mean','median','max']
grouped=data.groupby(['hotel','year'])
groupedStays=grouped[['stays_in_week_nights','stays_in_weekend_nights']].agg(aggMethods)
groupedStays
```

```python
#determine the number of repeated guests and derived a new column that calcualtes the percentage of booking cancellatons for both hotels
repeatedGuest=data.loc[data['is_repeated_guest']==1]
repeatedGuest=repeatedGuest[['is_repeated_guest']].groupby([data['hotel'],data['year']]).count()
repeatedGuest['percentage of repeated Guest (%)']=round((repeatedGuest['is_repeated_guest']/grouped['customers'].count())*100,2)
repeatedGuest=repeatedGuest.rename(columns={'is_repeated_guest':'number of repeated guests'})
repeatedGuest
```

```python
#Determine the number  of booking cancellations and derived a new column that calculates the percentage of booking cancellations for both hotels
print('Number of  bookings cancellations from City and Resort hotel based on year ')
cancelBooking=data.loc[data['is_canceled']==1]
cancelBooking=cancelBooking[['is_canceled']].groupby([data['hotel'],data['year']]).count()
cancelBooking['percentage of cancellation (%)']=round((cancelBooking['is_canceled']/grouped['customers'].count())*100,2)
cancelBooking=cancelBooking.rename(columns={'is_canceled':'Number of cancellation'})
cancelBooking
```

```python
#Determine customer's meal preferences throughout their stay at hotels
mealStats=data[['customers']].groupby([data['hotel'],data['meal']]).count()
mealStats
```

```python
#Determine the distribution chanel for customers to book their stay at the hotels
distributionData=data[['customers']].groupby([data['hotel'],data['customers']]).count()
distributionData=distributionData.rename(columns={'customers':'Number of customers'})
distributionData
```

```python
#droppinng unrelated columns from the dataframe
x=dataPrediction.drop(columns=['is_canceled','country','distribution_channel','date','reservation_status_date'])

#assigned a variable for attributets that will e predicted
y=dataPrediction['is_canceled']

#Set the train and the test dataset. Set the test dataset to the size 20% of the initial data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

#fit and predict the train model
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
predictions=model.predict(x_test)

#calcu;ate the accuray of the train dataset compated to the initial dataset
score=accuracy_score(y_test,predictions)
print(score)
print("The accuracy of the training dataset is ",(score*100),'%')

answer={'original':y_test, 'predictions':predictions}
frame=pd.DataFrame(answer)
frame
```
