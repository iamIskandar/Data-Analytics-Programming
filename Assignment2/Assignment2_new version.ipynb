```python
import numpy as np
import pandas as pd
```

```python
attendance=[]
with open('Attendance_RAW.txt') as line:
    lines=[int(f) for f in line]
    attendance.append(lines)
    
attendance_arr=np.array(attendance)
attendance_arr
```

```python
#row represent employee...
#column represent working days
attendance_arr=attendance_arr.reshape(14,15).T
attendance_arr
```

```python
attendance_arr.shape[0]
```

```python
def averageAttendance():
    mapping={}
    total=np.mean(attendance_arr,axis=0)
    i=0
    for x in range(1,attendance_arr.shape[1]+1):
        mapping[x]=round(total[i],2)
        i+=1
    return mapping
result=averageAttendance()
print(result)
```

```python
emp_detail=pd.read_csv('emp_detail.csv')
emp_detail[:]
```

```python
attendance=pd.DataFrame(attendance_arr)
attendance
```

```python
emp_attendance=pd.concat([emp_detail,attendance],axis=1)
emp_attendance
```

```python
emp_attendance['total']=np.sum(attendance_arr,axis=1)
emp_attendance
```

```python
emp_attendance['rate']=round(emp_attendance['total']/attendance_arr.shape[1],2)
emp_attendance
```

```python
def testing():
    final=[]
    for x in emp_attendance['rate']:
        if x>0.9:
            val='Good'
        elif x<0.6:
            val='Bad'
        else:
            val='Moderate'
        final.append(val)
    return final
emp_attendance['status']=testing()
emp_attendance
```
