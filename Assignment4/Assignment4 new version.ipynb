```python
import pandas as pd
```

```python
Customer=pd.read_csv('customer.csv')
Products=pd.read_csv('products.csv')
Sales=pd.read_csv('sales.csv')
```

```python
pd.DataFrame(Sales.groupby(['product','store'])['qty'].sum())
```

```python
jap=pd.merge(Sales,Products,on=['p_id','product'],how='left')
jap['totalPrice']=jap['qty']*jap['price']
jap
```
