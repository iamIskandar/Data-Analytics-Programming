**INSTRUCTIONS:**
1. Reads the data file in Python.
2. Creates a function in Python called: **calculate_daily_sale** that:
   - Able to calculate and return the sale of each item for every day.
   - You need to determine the data structure for the returned information, where the returned result should be able to be accessed to show the sale of exact item on a specified day. E.g.: 
      - ```result[“2020-09-25”]``` *will returns the sale for each item on that day.*
      - ```result[“2020-09-25”][“Item_B”]``` *will returns the sale of Item B on that day.*
3. Creates a function in Python called: **get_total_sale** that:
   - Able to combine the sale of a specific item throughout the period in the data file.
   - The function should receive ONE argument:
     - The result variable returned form **calculate_daily_sale**
   - The function should return the total sale of each item throughout the period.

````python
# NAME: MOHAMMAD IRFAN ISKANDAR BIN ZAINUDDIN
# MATRIC: A18CS0111
# SECTION: 01
# LECTURER: SUHAILA BINTI MOHAMAD YUSUF

# reading data from file
readFile = open("data_assignment_1.txt", "r")
tarikh = []
item = []
qty = []
price = []
# Slicing data from file and append into respective empty list
for x in readFile:
    tarikh.append(x[:x.find(" ")])
    item.append(x[x.find(" ")+1:x.find("_")+2])
    qty.append(x[18:20])
    price.append(x[20:x.find("\n")])
readFile.close()
print("\n\n")

# Creating empty dictionary and transfer element from list into myDict
myDict = {}
index = 0
def calculate_daily_sale():
    for x in tarikh:
        global index
        # checking if date is already the keys for myDict or not yet
        if x not in myDict.keys():
            # creating another dictionary (inside_my_dict)  and make item as the KEYS and total price as the VALUES and round the result to 2 decimal place
            inside_my_Dict = {item[index]: round((int(qty[index])*float(price[index])),2)}
            # inside_my_dict as the values for myDict
            myDict[x] = inside_my_Dict
        else:
            # if date already KEYS in myDict
            temp1 = myDict.get(x)
            nextItemtotal = temp1.get(item[index])  # (1.9*5)
            # check whtether next item is already in temp1 dictionary
            if item[index] in temp1:
                #accumulate the available item with the new item added
                nextItemtotal += round((int(qty[index])*float(price[index])),2)
                # update the valuyes 0f the exising item with varibale nextItemtotal
                temp1.update({item[index]: nextItemtotal})
            else:
                # add new item into temp1 dictionary and assign the total price as its values
                temp1[item[index]] = round((int(qty[index])*float(price[index])),2)
                # update the existing dictionary by adding new values to key X
                myDict.update({x: temp1})
        index += 1  # (index=1)
    return myDict

def get_total_sale(dic):
    total_A = total_B = total_C = total_D = total_E = total_F = 0.0
    for x in dic:
        # searching for values of items to accumulate the total
        item = dic.get(x)
        for y in item:
            if y == "Item_A":
                total_A += item.get(y)
            elif y == "Item_B":
                total_B += item.get(y)
            elif y == "Item_C":
                total_C += item.get(y)
            elif y == "Item_D":
                total_D += item.get(y)
            elif y == "Item_E":
                total_E += item.get(y)
            elif y == "Item_F":
                total_F += item.get(y)
            else:
                break
    # print the sale of each items throughout the period
    print("\nTotal Sale of Items throughout the period")
    print(40*"-")
    print("Total Sale for Item_A: RM {:.2f}".format(total_A))
    print("Total Sale for Item_B: RM {:.2f}".format(total_B))
    print("Total Sale for Item_C: RM {:.2f}".format(total_C))
    print("Total Sale for Item_D: RM {:.2f}".format(total_D))
    print("Total Sale for Item_E: RM {:.2f}".format(total_E))
    print("Total Sale for Item_F: RM {:.2f}".format(total_F))


# result now hold the entire dictionary
result=calculate_daily_sale()
#print the sale on "2020-09-23"
print(result["2020-09-23"])
print()
#print the sale for Item_B on "2020-09-25"
print(result["2020-09-25"]["Item_B"])
get_total_sale(result)
````
